```python
sns.set_style("darkgrid", {'grid.linestyle': ':', # "-", ":", "--"
                            'grid.linewidth': 1,
                            'axes.spines.top': True,
                            'axes.spines.right': False})
```

```python
#Matplotlib
# plt.scatter(tips["total_bill"], tips["tips"])
# plt.show()

#Seaborn
sns.scatterplot(data=tips, x = 'total_bill', y='tip')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='smoker',
    # size="size",
    # style="smoker",
    marker="*",
    color="27F549",
    s= 400
)

plt.show()
```

```python
sns.kdeplot(data=tips,
            x='total_bill',
            linewidth=6,
            linestyle=':',
            color= "#6CF527",
            )
plt.show()
```

```python
sns.kdeplot(
    data=tips,
    x='total_bill',
    hue='smoker',
    linewidth=1,
    linestyle=':',
    color= "#6CF527",
    fill=True, #ostini toldirish
    bw_adjust=0.5, #silliqlik darajasi
    alpha=0.5
)
```

```python
plt.figure(figsize=(10,8,))
sns.boxplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='sex',
    width=1, #quti kengligi
    gap =0.5 #hue orasidagi bo'shliq
)

plt.legend(
    title="Sex",
    fontsize=10,
    title_fontsize=15,
    loc='lower right'
)

plt.show()
```

```python
sns.violinplot(
    data=tips,
    x="day",
    y="total_bill",
    inner='box',
    # inner="quartile",
    # inner='quart',
    # inner='point',
    # inner=None,
    hue='sex',
    gap=0.1
)
plt.show()
```

```python
sns.stripplot(
    data=tips,
    x="day",
    y="total_bill",
    jitter=True,
    alpha=0.5,
    marker="*",
    s=15,
    color="#F5276C",
    linewidth=1
    )

plt.show()
```

```python
sns.regplot(data=tips, x='total_bill', y='tip', color="#AF3FDE",
            ci=95, #95% ishonch oraligi
            # ci=68, #torroq
            # ci=None) #zichroq
)
plt.show()
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hist')
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg')
```
