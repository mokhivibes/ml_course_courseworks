```python
oylar = [1, 2, 3, 4, 5, 6]
sotuv = [120, 150, 130, 170, 160, 200]

fig, ax = plt.subplots(1, 1, figsize =(10, 4), facecolor = "pink")
ax.plot(oylar, sotuv,
        color = "#fa348a",
        linestyle = ":",
        linewidth = 2,
        marker = "s"
        )

ax.set_title("Oylik Sotuv")
ax.set_xlabel("Oy")
ax.set_ylabel("Sotuv (mln)")
ax.grid(True)
plt.show()
```

```python
cars = ["Gentra", "Cobalt", "Malibu", "Tracker"]
sold = [530, 600, 200, 420]

fig, ax = plt.subplots(facecolor = "#b6b3ff")
ax.bar(cars, sold,
       width = 0.5,
       color = "#807aeb",
       edgecolor = "#070266")

ax.set_title("Sales of GM cars")
ax.set_xlabel("Cars")
ax.set_ylabel("Number of sold cars")
plt.show()
```

```python
kurslar = ["Python Asoslari", "Machine Learning", "Deep Learning", "Data Science"]
talabalar = [120, 95, 60, 85]

fig, ax = plt.subplots(facecolor = "#b6b3ff")
ax.barh(kurslar, talabalar,
       height = 0.5,
       color = "#807aeb",
       edgecolor = "#070266")

ax.set_title("Kurs bo'yicha talabalar soni")
ax.set_xlabel("Talabalar soni")
ax.grid(axis ="x")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.randint(40, 500, 80)
y = 3.8 * x + np.random.normal(0, 30, 80)

fig, ax = plt.subplots()
ax.scatter(x, y,
           s = 20, #nuqtalar yaxshi ko'rinsin
           c = "steelblue",
           alpha = 0.7)
ax.set_title("Uy maydoni vs Narxi")
ax.set_xlabel("Maydon (m²)")
ax.set_ylabel("Narx (mln so'm)")
ax.grid(True)
plt.show()
```
