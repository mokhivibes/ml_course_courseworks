```python
ism = input("Ism kiriting: ")
til = input("Tilni kiriting (uz, ru, eng): ")
til = til.lower()

def salomlash(ism, til = "uz"):
  if til == "uz":
    print(f"Salom, {ism.capitalize()}!")
  elif til == "eng":
    print(f"Hello, {ism.capitalize()}!")
  elif til == "ru":
    print(f"Привет, {ism.capitalize()}!")

salomlash(ism, til)
```
