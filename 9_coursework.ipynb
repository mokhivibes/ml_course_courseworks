```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
sns.set_theme()
```

```python
# --- 1. Normal taqsimor yaratish ----

mu = 170 #o'rtacha bo'y
sigma = 10 #standart og'ish

x = np.linspace(130, 210, 300) # x o'qi
y = stats.norm.pdf(x, mu, sigma) #PDF

#print(y)

# --- 2. Grafik ---

fig, axes = plt.subplots(1, 2, figsize = (14, 5))

#PDF
axes[0].plot(x, y, color="#2ECB9B", linewidth=2.5)
axes[0].fill_between(x, y, alpha=0.3, color="#2ECB9B")
axes[0].axvline(mu, color='red', linestyle='--')
axes[0].set_title('Normal PDF - Bo\'y taqsimoti')

#Simulyatsiya
samples = np.random.normal(mu, sigma, 1000)
axes[1].hist(samples, bins=30, color='#2ECB9B', ec="w")
axes[1].set_title('1000 ta tasodifiy namuna')
plt.tight_layout()
plt.show()
```

```python
# ── 1. Parametrlar ──────────────────────────
n = 10    # tanga necha marta tashlanadi
p = 0.5   # 'Chikka' chiqish ehtimoli

# ── 2. PMF — har k uchun ehtimollik ────────
k = np.arange(0, n+1)         # 0 dan 10 gacha

print(n)
print(k)
pmf = stats.binom.pmf(k, n, p)

print(f'P(5 Chikka) = {stats.binom.pmf(5,n,p):.3f}')  # 0.246

# ── 3. Grafik ───────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# PMF Bar chart
axes[0].bar(k, pmf, color='#3B82F6', edgecolor='white')
axes[0].set_title('Binomial PMF (n=10, p=0.5)')

# n=10, p=0.3 bilan solishtirish
pmf2 = stats.binom.pmf(k, n, 0.3)
axes[1].bar(k-0.2, pmf,  width=0.4, label='p=0.5', color='#3B82F6')
axes[1].bar(k+0.2, pmf2, width=0.4, label='p=0.3', color='#2ECB9B')
axes[1].legend() ; axes[1].set_title('p=0.5 vs p=0.3')
plt.tight_layout() ; plt.show()
```

```python
# ── 1. λ=3: soatiga 3 ta qo'ng'iroq ────────
lam = 3
k   = np.arange(0, 15)

pmf = stats.poisson.pmf(k, lam)

# Aniq hisob
print(f'P(0 qongiroq) = {stats.poisson.pmf(0,lam):.3f}')  # 0.050
print(f'P(3 qongiroq) = {stats.poisson.pmf(3,lam):.3f}')  # 0.224
print(f'P(5+ qongiroq) = {1-stats.poisson.cdf(4,lam):.3f}')# 0.185

# ── 2. Turli λ solishtiruv ──────────────────
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for ax, l in zip(axes, [1, 3, 8]):
    ax.bar(k, stats.poisson.pmf(k, l))
    ax.set_title(f'Poisson (λ={l})')
    ax.axvline(l, color='red', linestyle='--', label=f'λ={l}')
plt.tight_layout() ; plt.show()
```
