```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/raw_csv.csv")
```

```python
df.info()
```

```python
del_columns = ["imageUrl", "category", "lunchTime", "storeUrl"]
# df.columns = df.columns.str.strip()
df = df.drop(columns=del_columns, axis=1)
df.info()
```

```python
df["sold"] = (
    df['sold']
    .astype(str)
    .str.replace(r"[^\d]", "", regex=True)
    .replace("", "0")
    .astype(int)
)
```

```python
# Step 1 - clean and convert to float
df['shippingCost'] = (
    df['shippingCost']
    .astype(str)
    .str.replace(r"[^\d.]", "", regex=True)  # remove non-numeric except dot
    .replace("", "0")
    .astype(float)
)

# Step 2 - fill missing values with median
median_shipping = df['shippingCost'].median()
print(median_shipping)
df['shippingCost'] = df['shippingCost'].fillna(median_shipping)


df.info()
```

```python
before = len(df)
df = df.drop_duplicates(subset='id', keep="first")
after = len(df)

print(f"\n[Task 4] Duplicates removed: {before -after:,}")
print(f"                    Shape now: {df.shape}")
```
