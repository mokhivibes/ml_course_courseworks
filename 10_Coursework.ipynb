```python
import numpy as np
import pandas as pd

daromad = [1_000_000, 1_200_000, 1_100_000, 1_050_000, 1_150_000]

#Numpy bilan

print(np.mean(daromad))

#Pandas Series bilan
s = pd.Series(daromad)
print(s.mean())

#DataFrame  ustunida bilan
df = pd.DataFrame({'daromad': daromad})
print(df['daromad'].mean())
```

```python
from scipy import stats

baholar = [5, 5, 3, 6, 6, 5, 3, 6, 5, 6, 2] #eng kichkina sonni oladi
mode_result = stats.mode(baholar)
print(f"Eng ko'p takrorlangan son: {mode_result.mode} bo'ldi.")
print(f"{mode_result.mode} {mode_result.count} marta takrorlandi.")


ismlar = ['ali', 'valii', 'ali', 'valii', 'ali', 'valii']
df = pd.Series(ismlar)
print(df.mode())
```

```python
data = [2, 4, 4, 4, 5, 5, 7, 9]

#Populyatsiya dispersiyasi (ddof=0, standart)
print(np.var(data))

#Namunaviy dispersiya (ddof=1, Bessel korreksiyasi)
print(np.var(data, ddof=1))

#Pandas bilan (ddof=1 standart)
s = pd.Series(data)
print(s.var())
print(s.var(ddof=0))
```

```python
data = [2, 4, 4, 4, 5, 5, 7, 9]

print(np.std(data)) #populyatsiya
print(np.std(data, ddof=1)) #namunaviy
print()


#Pandas describe() - barcha statistika birdan
s = pd.Series(data)
print(s.std())
print()
print(s.describe())
```

```python
import matplotlib.pyplot as plt
np.random.seed(42)
data= np.random.normal(170, 10, 1000)

mean = np.mean(data)
std = np.std(data)

plt.figure(figsize=(10, 5))
plt.hist(data, bins=20, color='#2ECB9B', alpha=0.7, edgecolor='white')

plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean = {mean:.1f}")
plt.axvline(mean+std, color='blue', linestyle=':', linewidth=2, label=f"+1σ = {mean+std:.1f}")
plt.axvline(mean-std, color='blue', linestyle=':', linewidth=2, label=f"-1σ = {mean-std:.1f}")


plt.axvline(mean+(std*2), color='violet', linestyle='-', linewidth=2, label=f"+2σ = {mean+(std*2):.1f}")
plt.axvline(mean-(std*2), color='violet', linestyle='-', linewidth=2, label=f"-2σ = {mean-(std*2):.1f}")

plt.axvline(mean+(std*3), color='orange', linestyle='--', linewidth=2, label=f"+3σ = {mean+(std*3):.1f}")
plt.axvline(mean-(std*3), color='orange', linestyle='--', linewidth=2, label=f"-3σ = {mean-(std*3):.1f}")


plt.title('Talabalar bo\'yi taqsimoti')
plt.xlabel('Talabalar boyi')
plt.ylabel('Talabalar soni')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
df = sns.load_dataset('penguins').dropna()

col = df['bill_length_mm']
mean = col.mean()
std = col.std()

print('Mean: ', col.mean())
print('Median: ', col.median())
print('Std: ', col.std())
print('Var: ', col.var())

sns.histplot(col, bins=30, color='yellow',edgecolor= 'orange', lw=2, kde=True, line_kws={'color': 'orange', 'lw': 2})

plt.axvline(mean, color='red', linestyle='--', linewidth=2, label=f"Mean = {mean:.1f}")
plt.axvline(mean+std, color='blue', linestyle=':', linewidth=2, label=f"+1σ = {mean+std:.1f}")
plt.axvline(mean-std, color='blue', linestyle=':', linewidth=2, label=f"-1σ = {mean-std:.1f}")


plt.axvline(mean+(std*2), color='violet', linestyle='-', linewidth=2, label=f"+2σ = {mean+(std*2):.1f}")
plt.axvline(mean-(std*2), color='violet', linestyle='-', linewidth=2, label=f"-2σ = {mean-(std*2):.1f}")

plt.axvline(mean+(std*3), color='orange', linestyle='--', linewidth=2, label=f"+3σ = {mean+(std*3):.1f}")
plt.axvline(mean-(std*3), color='orange', linestyle='--', linewidth=2, label=f"-3σ = {mean-(std*3):.1f}")

plt.show()
```
