```python
#Attributelar

arr = np.array([[14, 7, 3], [7, 5, 6]])

print(arr.shape) # (2, 3) - 2 qator, 3 ustun
print(arr.ndim) # 2- ikki o'chamli
print(arr.size) # 6 - jami element soni
print(arr.dtype) # int64 - element turi
```

```python
arr = np.array([10, 20, 30, 40, 50])
arr[0] # 10
arr[-1] # 50
arr[1:4] # [20, 30, 40]
arr[::2] # [10, 30, 50]
np.random.seed(42)
arr2 = np.random.randint(0, 11, (3, 2))
print(arr2)
print(arr2[0:3, 1])
```

```python
#Statistik funksiyalar

arr = np.array([4, 5, 12, 14, 20])

print(np.sum(arr)) #umumiy qiymat
print(np.mean(arr)) # o'rtacha qiymat
print(np.std(arr)) # standart og'ish
print(np.min(arr)) # eng kichik qiymat
print(np.max(arr)) # eng katta qiymat
print(np.argmax(arr)) #maksimal element indeksi
```
