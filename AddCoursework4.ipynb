```python
#2-task
df["salary"]
df.dropna()["region"]
print(f"Mean of Salary: {df["salary"].mean()}")
print(f"Median salary: {df["salary"].median()}")
print(f"Maximum salary: {df["salary"].max()}")
print(f"Minimum salary: {df["salary"].min()}")
print(f"Standard deviation of salaries: {df["salary"].std()}")
print(f"Total sum of salaries: {df["salary"].sum()}")
```

```python
#task-4

print(f"The number of rows and columns: {df.shape}")
print(df.columns)
print(df.dtypes)
print(df.describe())
```

```python
#task-5
it_high = df[(df["department"]=="IT") & (df["salary"] > 8_000_000)]
print(it_high[["full_name","salary","city"]])
it_high["full_name"].shape
```

```python
#task-6
df["bonus_amount"] =(df["salary"] * df["bonus_percent"])/100
df["bonus_amount"] = df["bonus_amount"].round(-3)
df[(df["bonus_amount"] > 1_000_000)]
big_bonus = df[(df["bonus_amount"] > 1_000_000)]
print(big_bonus[["full_name","salary","bonus_percent","bonus_amount"]])
big_bonus["full_name"].shape
```

```python
big_bonus.to_csv("clean_employees.csv", index =False)
import matplotlib.pyplot as plt
import seaborn as sns

avg_salary = big_bonus.groupby("full_name")["salary"].mean()
# avg_salary.index = ismlar
# avg_salary.values = maoshlar

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(avg_salary.index, avg_salary.values)

ax.set_title("Ishchilar maoshi o'rtachasi")
ax.set_xlabel("Talabalar")
ax.set_ylabel("O'rtacha")
plt.xticks(rotation=45)  # ismlar qiyshiq bo'lsin
plt.show()
```
