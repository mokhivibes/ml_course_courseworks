```python
# Series = A Pandas 1-Dimensional labeled array that can hold any data type
            # Think of it like single column in a spreadsheet (1-Dimensional)


data = [100, 102, 104]

series = pd.Series(data, index=['a', 'b', 'c'])
print(series)

#Accessing array using label
print(series.loc['b'])
series.loc['a'] = 500
print(series)

#Accessing array using integer / index

print(series.iloc[2])


data = [100, 102, 104, 200, 202]
series = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'])
print(series[series < 200])
```

```python
calories = {"Day 1": 1750, "Day 2": 2100, "Day 3": 1700}

series = pd.Series(calories)
print(series.loc["Day 2"])

series.loc['Day 3'] = 2500
print(series)

print(series[series >= 2100])

```

```python
series = pd.Series(['Bulbasaur', 'Ivysaur', 'Venusaur', 'Charmander', 'Charmeleon', 'Charizard'], index = np.arange(1, 7))
print(series)
```

```python
#Data cleaning = the process of fixing/removing:
                 # incomplete, incorrect, or irrelent data.
                 #~75% of work done with Pandas is data cleaning

#1. Drop irrelevent columns
df.info()
df = df.drop(columns = [ 'Data_value' ])
```

```python
print(df.info())
```

```python
#2. Handle missing data
df =df.dropna(subset = ['Units']) #drops the row that there is not value in this units column

df = df.fillna({'Type': "None"})
```

```python
#3. Fix inconsistent values
df['Type'] = df['Type'].replace({'Single year': "SINGLE YEAR"})
print(df.to_string())
```

```python
#4, Standardize text

df['Cause'] = df['Cause'].str.lower()
print(df.to_string())
```

```python
#5. Fix data types

df['Upper_CI'] = df['Upper_CI'].astype(str)
print(df)
```

```python
#6. Remove duplicate values

df = df.drop_duplicates()
print(df.to_string())
```
